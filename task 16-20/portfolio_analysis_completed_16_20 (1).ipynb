### 1. Загрузка и обработка данных

```python
def preprocess_data(df: pd.DataFrame) -> pd.DataFrame:
    
    df.columns = df.columns.str.strip()
    df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
    df.set_index('date', inplace=True)

    df = df.replace('', np.nan).replace(' ', np.nan) \
            .dropna(how='all') \
            .dropna(axis=1, how='all')
    
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    return df
```

```python
def load_prices_data(file_path: str) -> pd.DataFrame:
    df = pd.read_csv(file_path, sep=';', decimal=',')

    return preprocess_data(df)
```

```python
stoks_prices_df = load_prices_data("prices_moex_new.csv")
```

Типичная логика интерпретации:
- при **unrestricted short** optimizer имеет максимум свободы, поэтому граница обычно наиболее гибкая;
- при **short limit 25%** возможности хеджирования уменьшаются;

```python
# Читаем отдельный файл с рыночным индексом IMOEX
# и считаем дневные логарифмические доходности индекса

try:
    imoex_df = pd.read_csv("imoex_prices.csv", sep=';', decimal=',')
except Exception:
    imoex_df = pd.read_csv("imoex_prices.csv")

# Если файл прочитался одной колонкой, пробуем еще раз
# в формате MOEX: разделитель ; и десятичная запятая
if imoex_df.shape[1] == 1:
    imoex_df = pd.read_csv("imoex_prices.csv", sep=';', decimal=',')

# Чистим названия колонок
imoex_df.columns = [col.strip() for col in imoex_df.columns]

# Оставляем только дату и закрытие индекса
imoex_df = imoex_df[["date", "IMOEX"]].copy()

# Приводим дату и значение индекса к нужному виду
imoex_df["date"] = pd.to_datetime(imoex_df["date"], dayfirst=True, errors="coerce")
imoex_df["IMOEX"] = pd.to_numeric(imoex_df["IMOEX"], errors="coerce")

# Удаляем пропуски, дубликаты и сортируем по дате
imoex_df = imoex_df.dropna(subset=["date", "IMOEX"])
imoex_df = imoex_df.drop_duplicates(subset=["date"], keep="last")
imoex_df = imoex_df.sort_values("date")

# Считаем логарифмические дневные доходности индекса
imoex_df["imoex_return"] = np.log(imoex_df["IMOEX"] / imoex_df["IMOEX"].shift(1))

# Делаем отдельный ряд доходностей индекса
imoex_returns = imoex_df.set_index("date")["imoex_return"].dropna()

# Оставляем только те даты индекса, которые есть в доходностях акций
proxy_market_returns = imoex_returns.loc[
    imoex_returns.index.intersection(returns.index)
].copy()

proxy_market_returns.name = "IMOEX"
market_label_16_20 = "IMOEX - official MOEX market index"

print("Рыночный ряд для пунктов 16-20 успешно подготовлен.")
print(f"Количество наблюдений по индексу: {len(proxy_market_returns)}")
print(f"Период индекса: {proxy_market_returns.index.min()} - {proxy_market_returns.index.max()}")
```
